```python
import pandas as pd
import numpy as np

# import data from csv file
air_y = pd.read_csv('aqi_daily_1980_to_2021.csv')
air_x = pd.read_csv('pollution_2000_2021.csv')
```

```python
# Sort the y data by matching it to the dates and locations in x data
air_y = air_y.sort_values(by=['Date', 'State Name', 'County Name'])
# Your code still doesn't match them becuase one is from 1980-2021 while the other is from 2000-2021
# You need to remove the 1980-1999 data from the y data
air_y = air_y[air_y['Date'] >= '2000-01-01']
air_y = air_y.reset_index(drop=True)

air_y.head()
```

```python
# make it change the column name from state name to state
air_y = air_y.rename(columns={'State Name': 'State'})
air_y = air_y.rename(columns={'County Name': 'County'})

air_y
```

```python
# make sure the x data is sorted by name and date
air_x = air_x.sort_values(by=['State', 'County', 'Date'])
air_x = air_x.reset_index(drop=True)

air_x.head()
```

```python
# Merge the two dataframes based on date, state, and county
air = pd.merge(air_x, air_y, on=['Date', 'State', 'County'])
air
```

```python
# Make the air_y the last 5 columns from the air dataframe
air_y = air.iloc[:, -5:]
# add the dates to the air_y dataframe
air_y['Date'] = air['Date']
air_y['State'] = air['State']
air_y
```

```python
# Write code that goes through the entire x data frame and searches to find the earliest date for each state

def find_earliest_date(state):
    earliest_date = air_x.loc[air_x['State'] == state, 'Date'].min()
    return earliest_date

# Make a list of all the states
states = air_x['State'].unique()

# Make a dictionary of all the earliest dates for each state
earliest_dates = {}
for state in states:
    earliest_dates[state] = find_earliest_date(state)
# sort the dictionary by date
earliest_dates = dict(sorted(earliest_dates.items(), key=lambda item: item[1]))
earliest_dates
```

```python
# Write code that deleates all data that is before 2016-03-14 for each state from the x data and y data
for state in earliest_dates:
    air_x = air_x.drop(air_x[(air_x['State'] == state) & (air_x['Date'] < earliest_dates['Vermont'])].index)
    air_y = air_y.drop(air_y[(air_y['State'] == state) & (air_y['Date'] < earliest_dates['Vermont'])].index)

air_y
```

```python
# Write code that goes through the entire x data frame and searches to find the earliest date for each state

def find_latest_date(state):
    earliest_date = air_x.loc[air_x['State'] == state, 'Date'].max()
    return earliest_date

# Make a list of all the states
states = air_x['State'].unique()

# Make a dictionary of all the earliest dates for each state
latest_dates = {}
for state in states:
    latest_dates[state] = find_latest_date(state)
# sort the dictionary by date
latest_dates = dict(sorted(latest_dates.items(), key=lambda item: item[1]))
latest_dates
```
